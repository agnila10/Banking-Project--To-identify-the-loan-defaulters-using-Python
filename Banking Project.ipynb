Missing Values Inputation

```python
app_msng_rmvd.isnull().sum().sort_values() * 100/app_msng_rmvd.shape[0]
```

```python
app_msng_rmvd['CNT_FAM_MEMBERS'] = app_msng_rmvd['CNT_FAM_MEMBERS'].fillna(app_msng_rmvd['CNT_FAM_MEMBERS'].mode()[0])
app_msng_rmvd['CNT_FAM_MEMBERS'].isnull().sum()
```

```python
app_msng_rmvd['OCCUPATION_TYPE'] = app_msng_rmvd['OCCUPATION_TYPE'].fillna(app_msng_rmvd['CNT_FAM_MEMBERS'].mode()[0])
app_msng_rmvd['OCCUPATION_TYPE'].isnull().sum()
```

```python
app_msng_rmvd['AMT_ANNUITY'] = app_msng_rmvd['AMT_ANNUITY'].fillna(app_msng_rmvd['AMT_ANNUITY'].mean())
app_msng_rmvd['AMT_ANNUITY'].isnull().sum()
```

```python
amt_req_col = []

for col in app_msng_rmvd.columns:
    if col.startswith("AMT_REQ_CREDIT_BUREAU"):
        amt_req_col.append(col)

amt_req_col
```

```python
for col in amt_req_col:
    app_msng_rmvd[col] = app_msng_rmvd[col].fillna(app_msng_rmvd[col].median())
```

```python
app_msng_rmvd['AMT_GOODS_PRICE'] = app_msng_rmvd['AMT_GOODS_PRICE'].fillna(app_msng_rmvd['AMT_GOODS_PRICE'].median())
app_msng_rmvd['AMT_GOODS_PRICE'].isnull().sum()
```
